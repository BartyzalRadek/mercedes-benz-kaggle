```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
# import seaborn as sns
# from sklearn import preprocessing
# color = sns.color_palette()

%matplotlib inline

pd.options.mode.chained_assignment = None  # default='warn'
pd.options.display.max_columns = 999
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")
print("Train shape : ", train_df.shape)
print("Test shape : ", test_df.shape)
```

```python
plt.scatter(range(train_df.shape[0]), np.sort(train_df.y.values))
plt.title('All y values, sorted')
plt.xlabel('index', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.show()
```

```python
print("Row with outlier value = {}".format(train_df['y'].argmax()))

outlier_row_index = train_df['y'].argmax()
train_df = train_df.drop(train_df.index[outlier_row_index])
plt.scatter(range(train_df.shape[0]), np.sort(train_df.y.values))
plt.title('All y values, sorted')
plt.xlabel('index', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.show()
```

```python
plt.scatter(train_df.ID.values, train_df.y.values)
plt.title('Is y related to ID?')
plt.xlabel('ID', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.show()
```

### Correlation

```python
calc_corr = train_df.corr()
calc_corr.head()
```

```python
calc_corr_filled = calc_corr.fillna(value=0)
calc_corr_filled.head()
```

```python
fig = plt.figure()
plt.imshow(calc_corr_filled.values, cmap='hot', interpolation='nearest')
plt.savefig('test.png', bbox_inches='tight', dpi=1200)
plt.show()
```

```python
plt.pcolor(calc_corr_filled.values)
```

### Correlation of columns to y:

```python
# Set size of plot
# %pylab inline
# pylab.rcParams['figure.figsize'] = (20, 10)

# Do not count ID and y itself
column_corr = calc_corr_filled['y'].values[2:]
columns = np.array(list(calc_corr_filled)[2:])
column_corr = np.sort(column_corr)
columns = columns[np.argsort(column_corr)]

fig, ax = plt.subplots()
plt.scatter(range(len(columns)), column_corr)
plt.title('Correlation to y')
ax.set_xticks(range(len(columns)))
ax.set_xticklabels(columns, rotation=90)
plt.ylabel('Correlation to y', fontsize=12)
plt.show()
```
